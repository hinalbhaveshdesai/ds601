```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
#sns.scatterplot(data = diamonds, x="carat", y="price", )
sns.displot(data = diamonds, x="carat", y="price", hue = "clarity", height= 8, aspect= 1.7)
"""
From the graph we figure that maximum diamonds are under the 2 carat mark and hence it is expected that their price should be low.
Which is quite evident from the figure. As the carat increases, the price increases.
"""
```

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True)
plot.ax.tick_params(labelrotation = 90)

"""
We see that the maximum diadomds color and clarity is accumulated at the centre.
This means that maximum diamonds have the 'G' color and have 'VS2' clarity
Very few diamonds are of the I1 quality.
"""
```
