```python
### DON'T CHANGE THIS CELL
import os
import json
import re
import pandas as pd

class FileProcessor:
    
    def __init__(self, path, file_types=['.ipynb']):
        self.files = self.find_all_files(path, file_types)
        print(f'{len(self.files)} file(s) found...')

    
    def find_all_files(self, path, file_types):
        paths = []
        for root, dirs, files in os.walk(path):
            for file in files:
                if any([file.endswith(ft) for ft in file_types]):
                    path = os.path.join(root, file)
                    # print(path)
                    paths.append(path)
        return paths
    
    def read_file(self, file_path, encoding='utf8'):
        with open(file_path,encoding=encoding) as f:
            return f.read()
    
    def start(self):
        """
        Entry point for the class
        """
        for file in self.files:
            self.process_file(file)
        
    def process_file(self, file_path):
        """
        abstract method
        """
        raise NotImplemented('process_file method not implemented')
        
    def __repr__(self):
        return f'{self.files}'
```

```python
### your code starts here

class ModuleLister(FileProcessor):
    def __init__(self):
        super().__init__(path='../', file_types=['.ipynb'])
    
    def process_file(self, file_path):
        try:
            dict1= {'FileName':[],'Module':[]}
            content = self.read_file(file_path)
            startpattern = re.compile(r'import ')
            endpattern = re.compile(r'\s|\t|\\n')
            point = [m.end() for m in startpattern.finditer(content)]
            word = []
            for item in point:     
                word.append((content[item:item+endpattern.search(content[item:]).start()]).rstrip('\"'))
            word = list(set(word))
            if '*' in word:
                word.remove('*')
            dict1['FileName'].append(file_path)
            dict1['Module'].append(word)
            df = pd.DataFrame.from_dict(dict1)
            print(df)
    
        except Exception as e:
            print('--- ERRROR', file_path, str(e))
    pass
```

```python
processor.start()
```
